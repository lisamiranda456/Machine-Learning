```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from mpl_toolkits.mplot3d import Axes3D
```

```python
data=pd.read_csv('student.csv')
```

```python
data.head()
```

```python
y=data["Writing"].values
```

```python
x=data[["Math","Reading"]].values
x
```

```python
math=data["Math"].values
read=data["Reading"].values
write=data["Writing"].values
```

```python
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(math,read,write,color="red")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
reg=LinearRegression()
reg=reg.fit(x_train,y_train)
```

```python
print('a1,a2=',reg.coef_)
print('a0=',reg.intercept_)
```

```python
#predicting the values for the test set
y_pred=reg.predict(x_test)
print(y_pred)
print(y_test)
```

```python
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(x_train[:,0],x_train[:,1],y_train,color="red")

plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print('RMSE=',rmse)
```

```python
r2=r2_score(y_test,y_pred)
print('Regressor score is=',r2)
```

```python
#residual plot
residue=y_test-y_pred
x=[i for i in range(1,len(y_pred)+1)]
plt.scatter(x,residue,color="red")
plt.plot(x,[0]*len(y_pred),color="Black")
plt.title("Residual Plot")
```
